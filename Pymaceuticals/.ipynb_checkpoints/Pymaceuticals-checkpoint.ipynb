```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_trendline(x,y,line_format):
    z = np.polyfit(x,y,1)
    p = np.poly1d(z)
    plt.plot(x,p(x),line_format)
```

```python
file1 = "data/mouse_drug_data.csv"
file2 = "data/clinicaltrial_data.csv"
mousedata = pd.read_csv(file1)
clinicaldata = pd.read_csv(file2)
mousedata.head()
```

```python
clinicaldata.head()
```

```python
merged_data = pd.merge(clinicaldata,mousedata, on = "Mouse ID")
merged_data.head()
```

```python
merged_data["Drug"].unique()
```

```python
#Tumor Response to Treatment (t.vol./time pt.)
tumor_resp_reord = merged_data.reset_index().groupby(["Timepoint","Drug"])["Tumor Volume (mm3)"].aggregate("mean").unstack()
tumor_resp_reord.head()
```

```python
tumor_resp_reord = tumor_resp_reord[["Capomulin","Infubinol","Ketapril","Placebo"]]
tumor_resp_reord.head(10)
```

```python
timept = tumor_resp_reord.index
t_vol_capo = tumor_resp_reord["Capomulin"]
t_vol_infu = tumor_resp_reord["Infubinol"]
t_vol_keta = tumor_resp_reord["Ketapril"]
t_vol_plac = tumor_resp_reord["Placebo"]
plt.scatter(timept,t_vol_capo, marker="o", facecolors="red", edgecolors="red", label="Capomulin")
plt.scatter(timept,t_vol_infu, marker="v", facecolors="blue", edgecolors="blue", label="Infubinol")
plt.scatter(timept,t_vol_keta, marker="d", facecolors="green", edgecolors="green", label="Ketapril")
plt.scatter(timept,t_vol_plac, marker="s", facecolors="aqua", edgecolors="aqua",label="Placebo")
plt.ylim(35,75)
plt.xlim(-5,50)
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.legend(loc="best")
plt.grid(b=None, which="major", axis="y")

inputs = [(t_vol_capo, "r-"),(t_vol_infu, "b-"),(t_vol_keta, "g-"), (t_vol_plac, "c-")]
for pair in inputs:
    y,line_format = pair
    plot_trendline(timept, y, line_format)

plt.show()
#error bars(standard error)
```

```python
#Metastatic Response to Treatment
meta_reord = merged_data.reset_index().groupby(["Timepoint","Drug"])["Metastatic Sites"].aggregate("mean").unstack()
meta_reord = meta_reord[["Capomulin","Infubinol","Ketapril","Placebo"]]
meta_reord.head(10)
```

```python
timept_m = meta_reord.index
meta_capo = meta_reord["Capomulin"]
meta_infu = meta_reord["Infubinol"]
meta_keta = meta_reord["Ketapril"]
meta_plac = meta_reord["Placebo"]
plt.scatter(timept_m,meta_capo, marker="o", facecolors="red", edgecolors="red", label="Capomulin")
plt.scatter(timept_m,meta_infu, marker="v", facecolors="blue", edgecolors="blue", label="Infubinol")
plt.scatter(timept_m,meta_keta, marker="d", facecolors="green", edgecolors="green", label="Ketapril")
plt.scatter(timept_m,meta_plac, marker="s", facecolors="aqua", edgecolors="aqua",label="Placebo")
plt.ylim(-1,4)
plt.xlim(-5,50)
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.legend(loc="best")
plt.grid(b=None, which="major", axis="y")

inputs = [(meta_capo, "r-"),(meta_infu, "b-"),(meta_keta, "g-"), (meta_plac, "c-")]
for pair in inputs:
    y,line_format = pair
    plot_trendline(timept_m, y, line_format)
    
plt.show()
#error bars(standard error)
```

```python
#Survival Rates
surv_rate = merged_data[["Drug","Timepoint","Mouse ID"]]
surv_rate_4 = surv_rate.groupby(["Drug","Timepoint"])
surv_count = surv_rate_4.count()
surv_count = surv_count.rename(columns={"Mouse ID":"Mouse Count"})
surv_count_4 = surv_count.drop(["Naftisol", "Stelasyn","Ramicane","Propriva","Zoniferol","Ceftamin"])
surv_count_4.head()
surv_count_reord = surv_count_4.groupby(["Timepoint","Drug"])["Mouse Count"].aggregate("first").unstack()
surv_count_reord.head()
#need survival rate as percent
```

```python
timept_s = surv_count_reord.index
surv_capo = surv_count_reord["Capomulin"]
surv_infu = surv_count_reord["Infubinol"]
surv_keta = surv_count_reord["Ketapril"]
surv_plac = surv_count_reord["Placebo"]
plt.scatter(timept_s,surv_capo, marker="o", facecolors="red", edgecolors="red", label="Capomulin")
plt.scatter(timept_s,surv_infu, marker="v", facecolors="blue", edgecolors="blue", label="Infubinol")
plt.scatter(timept_s,surv_keta, marker="d", facecolors="green", edgecolors="green", label="Ketapril")
plt.scatter(timept_s,surv_plac, marker="s", facecolors="aqua", edgecolors="aqua",label="Placebo")
plt.ylim(7,27)
plt.xlim(-5,50)
plt.title("Subject Survival During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Survival Rates")
plt.legend(loc="best")
plt.grid(b=None, which="major", axis="y")

inputs = [(surv_capo, "r-"),(surv_infu, "b-"),(surv_keta, "g-"), (surv_plac, "c-")]
for pair in inputs:
    y,line_format = pair
    plot_trendline(timept_s, y, line_format)
    
plt.show()
# #error bars(standard error)
```
